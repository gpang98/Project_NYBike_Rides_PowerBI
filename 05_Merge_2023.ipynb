```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Dependencies
from pathlib import Path
```

```python
# # If you want the column to be interpreted as integers
# #dtypes = {'column5': int, 'column7': int}

# # If you want the column to be interpreted as floating-point numbers
# #dtypes = {'column5': float, 'column7': float}

# # If you want the column to be interpreted as strings
# dtypes = {'column5': str, 'column7': str}

# # If you want to handle mixed types, you can use object
dtypes = {'column5': object, 'column7': object}

# Read the CSV file with specified dtypes
df1 = pd.read_csv('Resources/01_202302-citibike-tripdata_winter_edit.csv', dtype=dtypes, low_memory=False)
#df2 = pd.read_csv('Resources/02_202305-citibike-tripdata_spring_edit.csv', dtype=dtypes, low_memory=False)
df3 = pd.read_csv('Resources/03_202308-citibike-tripdata_summer_edit.csv', dtype=dtypes, low_memory=False)
#df4 = pd.read_csv('Resources/04_202310-citibike-tripdata_fall_edit.csv', dtype=dtypes, low_memory=False)
```

## Combined Winter (Feb) and Summer (Aug) 2023 data only

```python
# Concatenate along the rows (axis=0)
combined_df1 = pd.concat([df1, df3], axis=0)

# Resetting index of the combined DataFrame
combined_df1.reset_index(drop=True, inplace=True)
combined_df1.head()
```

```python
combined_df1.info()
```

```python
combined_df1.describe().round(2)
```

```python
# Change data types to date
combined_df1[['started_at']] = combined_df1[['started_at']].apply(pd.to_datetime)
combined_df1[['ended_at']] = combined_df1[['ended_at']].apply(pd.to_datetime)
```

```python
print(combined_df1.info())
combined_df1.describe().round(2)
```

```python
pd.DataFrame.from_records([(col, combined_df1[col].nunique(), combined_df1[col].dtype, combined_df1[col].memory_usage(deep=True) )  for col in combined_df1.columns],
  columns=['Column Name', 'Unique', 'Data Type','Memory Usage'])
```

```python
# To save to a CSV file named 'output.csv'
combined_df1.to_csv('Resources/Merged_citibike_tripdata_Feb_Aug2023.csv', index=False)
```
