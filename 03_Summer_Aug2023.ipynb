```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Dependencies
from pathlib import Path
```

```python
# # If you want the column to be interpreted as integers
# #dtypes = {'column5': int, 'column7': int}

# # If you want the column to be interpreted as floating-point numbers
# #dtypes = {'column5': float, 'column7': float}

# # If you want the column to be interpreted as strings
# dtypes = {'column5': str, 'column7': str}

# # If you want to handle mixed types, you can use object
dtypes = {'column5': object, 'column7': object}

# Read the CSV file with specified dtypes
df1 = pd.read_csv('Resources/03_202308-citibike-tripdata_summer.csv', dtype=dtypes, low_memory=False)
```

```python
df1.head()
```

```python
df1.info()
```

```python
df1.describe().round(2)
```

```python
# Change data types to date
df1[['started_at']] = df1[['started_at']].apply(pd.to_datetime)
df1[['ended_at']] = df1[['ended_at']].apply(pd.to_datetime)
```

```python
print(df1.info())
df1.describe().round(2)
```

```python
pd.DataFrame.from_records([(col, df1[col].nunique(), df1[col].dtype, df1[col].memory_usage(deep=True) )  for col in df1.columns],
  columns=['Column Name', 'Unique', 'Data Type','Memory Usage'])
```

## Reduce the dataset to 20% randomly

```python
# Sample 30% of the data
df1_reduced = df1.sample(frac=0.2, random_state=42)  # Adjust random_state for reproducibility
```

```python
print(df1_reduced.info())
df1_reduced.describe().round(2)
```

```python
pd.DataFrame.from_records([(col, df1_reduced[col].nunique(), df1_reduced[col].dtype, df1_reduced[col].memory_usage(deep=True) )  for col in df1_reduced.columns],
  columns=['Column Name', 'Unique', 'Data Type','Memory Usage'])
```
